```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pymongo
import pprint


        
```

```python
course_cluster_uri = 'mongodb://admin:1Pfhr39Hdi4@ds119060.mlab.com:19060/pat_doc_recipedb'
```

```python
course_client = pymongo.MongoClient(course_cluster_uri)
recipes = course_client['pat_doc_recipedb']['recipes']
```

```python
query = {
    "upvotes": {"$exists": True}
}

projection = {
    "upvotes" : 1,
    "recipe_name": 1,
    '_id': 0
}
```

```python
rm = list(recipes.find(query, projection))
```

```python
pprint.pprint(rm[0])
```

```python
upvotes = [recipe['upvotes'] for recipe in rm]
```

```python
print(upvotes)
```

```python
recipe_names = [recipe["recipe_name"] for recipe in rm]
```

```python
print (recipe_names)
```

```python
plt.clf()
fig, ax = plt.subplots()
ax.bar(recipe_names, upvotes, alpha=0.5)

plt.title("UpVotes per Recipe")
plt.xlabel('Recipe Names')
plt.ylabel("UpVotes")

plt.show
```
